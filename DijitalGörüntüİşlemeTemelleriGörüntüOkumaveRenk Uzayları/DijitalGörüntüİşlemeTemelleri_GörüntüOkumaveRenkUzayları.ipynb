```python
# RGB (Red-Green-Blue), bilgisayarda renkle 0 ile 255 arasında temsil edilir
# Yani, renkler gösterilirken uint8 veri tipi kullanılır.
# uint8 demek unsignedint8 demektir. İşaretsiz 8 bitlik değer:
# 0 ile 255 arasında (her ikisi de dahil) 256 adet tamsayı vardır.
```

```python
# RGB ile Gray level (gri seviye) görüntü arasında sadece kanal ve gösterim
# farkı vardır. Kanal farkı, gri seviye görüntüde 1 kanal vardır, RGB'de ise 3 tane.
# RGB, renkli görüntü gösterirken, gri seviye sadece siyah veya beyaz tonlarında
# renkleri gösterir.
```

```python
# cv2 ile BGR kanal sıralamasında okunan görüntüyü RGB kanal sıralamasına dönüştürüyoruz

img_rgb_1 = cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb_1)
plt.axis("off")
```

```python
img_gray_1 = cv2.cvtColor(img_rgb_1,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_gray_1,cmap="gray")
plt.axis("off")
```

```python
print(img_rgb_1.shape)
print(img_gray_1.shape)
```

```python
print(f"RGB görüntünün piksel değeri = {img_rgb_1[500,500]}")
print(f"Gri görüntünün piksel değeri = {img_gray_1[500,500]}")
```

```python
# Görüntünün Yüksekliği satır sayısıyla doğru orantılıdır
# Görüntünün genişliği sütun sayısıyla doğru orantılıdır
```

```python
# Tek bir plot içerisinde alt plotlarla (subplot) birden fazla görüntünün
# gösterimi

fig = plt.figure(figsize=(10,7))

#plt.subplot(satır_sayısı,sütun_sayısı,o anki gösterilecek görüntü)
plt.subplot(1,2,1)
plt.imshow(img_rgb_1)
plt.axis("off")
plt.title("RGB Image")

plt.subplot(1,2,2)
plt.imshow(img_gray_1,cmap="gray")
plt.axis("off")
plt.title("Gray Level Image")


plt.tight_layout()
plt.savefig("My_image_with_subplot.png",dpi=300)
plt.show()
```
